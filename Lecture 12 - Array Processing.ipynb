```python
A=np.array([1,2,3,4])
B=np.array([100,200,300,400])
```

```python
condition=np.array([True,True,False,False])
```

```python
#Array comprehension
answer=[A_val if cond else B_val for A_val,B_val,cond in zip(A,B,condition)]
```

```python
answer
```

```python
answer2=np.where(condition,A,B) #short form of above code
```

```python
answer2
```
